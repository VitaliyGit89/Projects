–ó–∞–∫–∞–∑—á–∏–∫ —Ö–æ—á–µ—Ç –ø–æ–Ω—è—Ç—å, –∑–∞–≤–∏—Å–∏—Ç –ª–∏ –ø–æ–ª–Ω–æ—Ç–∞ —Å–≤–µ–¥–µ–Ω–∏–π –æ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–∞—Ö (–Ω–∞–ø—Ä–∏–º–µ—Ä, –æ–± –∏—Ö –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏) –æ—Ç —Ä–∞–∑–º–µ—Ä–∞ –∫–æ–º–ø–∞–Ω–∏–π.

- –û—Ü–µ–Ω–∏—Ç–µ, –Ω–∞—Å–∫–æ–ª—å–∫–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –ø–æ–ª–Ω–∞. –ò—Å–ø–æ–ª—å–∑—É—è –¥–∞—Ç–∞—Å–µ—Ç—ã `people.csv` –∏ `education.csv`, —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –≤—Å–µ –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä—É–ø–ø –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –∏ –æ—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω—é—é –¥–æ–ª—é —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –±–µ–∑ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ –≤ –∫–∞–∂–¥–æ–π –∏–∑ –≥—Ä—É–ø–ø. –û–±–æ—Å–Ω—É–π—Ç–µ –≤—ã–±—Ä–∞–Ω–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã –≥—Ä—É–ø–ø.
- –û—Ü–µ–Ω–∏—Ç–µ, –≤–æ–∑–º–æ–∂–Ω–æ –ª–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–∏—Ç—å –∫ —ç—Ç–∏–º —Ç–∞–±–ª–∏—Ü–∞–º –µ—â—ë –∏ —Ç–∞–±–ª–∏—Ü—É `degrees.csv`.

```python
# –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ç–∞–±–ª–∏—Ü—ã people.csv –∏ education.csv
ppl_and_edu_df = pd.merge(people_df, education_df, left_on='id', right_on='person_id', how='left')
ppl_and_edu_df.info()
ppl_and_edu_df.head()
```

```python
# –°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –∫–æ–ª-–≤—É —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤ –∫–æ–º–ø–∞–Ω–∏—è—Ö
ppl_and_edu_df_grouped = ppl_and_edu_df.groupby('company_id')[['id_x', 'instituition']].count().reset_index()

# –î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü no_edu —Å –¥–æ–ª—è–º–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –∫–æ–º–ø–∞–Ω–∏–π —Å –±–µ–∑ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ–± –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏
ppl_and_edu_df_grouped['no_edu'] = (1 - ppl_and_edu_df_grouped['instituition'] / ppl_and_edu_df_grouped['id_x']) * 100

ppl_and_edu_df_grouped.head()
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º –æ–ø–∏—Å–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –∫–æ–ª-–≤–æ–º —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤, —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –Ω–∞ –∫–∞–∫–∏–µ –≥—Ä—É–ø–ø—ã –∏—Ö –ø–æ–¥–µ–ª–∏—Ç—å. 
# –¢–∞–∫–∂–µ –æ—Ç–º–µ—Ç–∏–º 99-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å, —á—Ç–æ–±—ã –∏—Å–∫–ª—é—á–∏—Ç—å –∞–Ω–æ–º–∞–ª–∏–∏  
ppl_and_edu_df_grouped['id_x'].describe(percentiles=[0.75, 0.99])
```

```python
# –ü–æ–¥–µ–ª–∏–º –Ω–∞ 4 –≥—Ä—É–ø–ø—ã: 
# small - –Ω–µ –±–æ–ª–µ–µ 15 —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤, medium - –Ω–µ –±–æ–ª–µ–µ 100, big - –Ω–µ –±–æ–ª–µ–µ 250, huge - –±–æ–ª–µ–µ 250.
ppl_and_edu_df_grouped['rank']  = pd.cut(ppl_and_edu_df_grouped['id_x'], 
                                  bins=[0, 15, 100, 250, float('inf')], labels=['small', 'medium', 'big', 'huge'])

ppl_and_edu_df_grouped.head()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=ppl_and_edu_df_grouped, x='id_x')
plt.show()
```

```python
# –ö –¥–∞—Ç–∞—Å–µ—Ç—É df_company_boss –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–∏–º acquisition_filtered –ø–æ –ø–æ–ª—é acquired_company_id (id –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–º–ø–∞–Ω–∏–π)
df_with_sold_companies = df_company_boss.merge(acquisition_filtered, left_on='company_id', right_on='acquired_company_id')

# –û—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞—Ç–∞—Å–µ—Ç –∏ –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å –Ω–µ–Ω—É–ª–µ–≤—ã–º –æ–±—â–∏–º –æ–±—ä—ë–º–æ–º —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è (funding_total)
df_with_sold_companies = df_with_sold_companies[df_with_sold_companies['funding_total'] != 0]
df_with_sold_companies.head()
df_with_sold_companies['funding_total'].describe(percentiles=[0.01, 0.1, 0.25, 0.5, 0.75, 0.99])
```

```python
# –†–∞—Å—á—ë—Ç –∫–≤–∞–Ω—Ç–∏–ª–µ–π
Q1 = df_with_sold_companies['funding_total'].quantile(0.25)
Q3 = df_with_sold_companies['funding_total'].quantile(0.75)
IQR = Q3 - Q1

# –ì—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (25-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å): {Q1}")
print(f"Q3 (75-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—å): {Q3}")
print(f"IQR: {IQR}")
print(f"–ù–∏–∂–Ω–∏–π –ø–æ—Ä–æ–≥ –≤—ã–±—Ä–æ—Å–æ–≤: {lower_bound}")
print(f"–í–µ—Ä—Ö–Ω–∏–π –ø–æ—Ä–æ–≥ –≤—ã–±—Ä–æ—Å–æ–≤: {upper_bound}")
```
