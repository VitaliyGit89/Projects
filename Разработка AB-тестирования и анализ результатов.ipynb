#### 2.1 –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –Ω—É–ª–µ–≤–æ–π –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑

–ü–µ—Ä–µ–¥ —Ç–µ–º –∫–∞–∫ –ø—Ä–æ–≤–æ–¥–∏—Ç—å –ê/B-—Ç–µ—Å—Ç, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –Ω—É–ª–µ–≤—É—é –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—ã. –ù–∞–ø–æ–º–Ω–∏–º –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ: –∫–æ–º–∞–Ω–¥–∞ —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–æ–≤ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ç–µ–ª—å–Ω—ã—Ö —Å–∏—Å—Ç–µ–º —Å–æ–∑–¥–∞–ª–∞ –Ω–æ–≤—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º, –∫–æ—Ç–æ—Ä—ã–π, –ø–æ –∏—Ö –º–Ω–µ–Ω–∏—é, –±—É–¥–µ—Ç –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å –±–æ–ª–µ–µ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–π –∫–æ–Ω—Ç–µ–Ω—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

#### 2.2. –†–∞—Å—á—ë—Ç —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏
–í —Ä–∞–º–∫–∞—Ö –∫—É—Ä—Å–∞ –≤—ã —É–∂–µ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–ª–∏ —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–∫–∏ –∏  –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –¥–ª—è —ç—Ç–æ–≥–æ –æ–Ω–ª–∞–π–Ω-–∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä. –í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –≥–æ—Ç–æ–≤—ã–º –∫–æ–¥–æ–º –∏ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ –¥–ª—è –≤–∞—à–µ–≥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–î–ª—è —ç—Ç–æ–≥–æ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –≤ –∫–æ–¥–µ –Ω–∏–∂–µ —Å–ª–µ–¥—É—é—â–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:

- –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ‚Äî 0.05.

- –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞ ‚Äî 0.2.

- –ú–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞.

- –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –¥–µ—Ç–µ–∫—Ç–∏—Ä—É–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç, –∏–ª–∏ MDE, ‚Äî 3%. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∑–¥–µ—Å—å –Ω—É–∂–Ω–æ —É–∫–∞–∑–∞—Ç—å –¥–µ—Å—è—Ç–∏—á–Ω—É—é –¥—Ä–æ–±—å, –∞ –Ω–µ –ø—Ä–æ—Ü–µ–Ω—Ç.

–ü—Ä–∏ —Ä–∞—Å—á—ë—Ç–µ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–µ—Ç–æ–¥ `solve_power()` –∏–∑ –∫–ª–∞—Å—Å–∞ `power.NormalIndPower` –º–æ–¥—É–ª—è `statsmodels.stats`.

–ó–∞–ø—É—Å—Ç–∏—Ç–µ —è—á–µ–π–∫—É –∏ –∏–∑—É—á–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

````python
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize

# –ó–∞–¥–∞—ë–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
beta = 0.2  # –û—à–∏–±–∫–∞ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞, —á–∞—Å—Ç–æ 1 - –º–æ—â–Ω–æ—Å—Ç—å
power = 1 - beta  # –ú–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞
p = 0.3 # –ë–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –¥–æ–ª–∏
mde = 0.03 * p  # –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –¥–µ—Ç–µ–∫—Ç–∏—Ä—É–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç
effect_size = proportion_effectsize(p, p + mde)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Å NormalIndPower
power_analysis = NormalIndPower()

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1 # –†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫
)

print(f"–ù–µ–æ–±—Ö–æ–¥–∏–º—ã–π —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã: {int(sample_size)}")
````

`power` –º–æ–∂–Ω–æ —Ç–∞–∫–∂–µ –∑–∞–¥–∞—Ç—å —á–µ—Ä–µ–∑ `beta`: `power = 1 - beta`

#### 2.3. –†–∞—Å—á—ë—Ç –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ A/B-—Ç–µ—Å—Ç–∞

–û—Å–æ–±–µ–Ω–Ω–æ –≤–∞–∂–Ω–æ, —á—Ç–æ –±—ã–ª–∞ –ø—Ä–æ–≤–µ—Ä–µ–Ω–∞ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –∏ —á—Ç–æ –æ–Ω–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –ø—Ä–µ–¥–µ–ª–∞—Ö.

### 4. –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ A/B-—Ç–µ—Å—Ç–∞

A/B-—Ç–µ—Å—Ç –∑–∞–≤–µ—Ä—à—ë–Ω, –∏ —É –≤–∞—Å –µ—Å—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∑–∞ –≤—Å–µ –¥–Ω–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ —É–±–µ–¥–∏—Ç—å—Å—è –≤ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —Ç–µ—Å—Ç–∞ –∏ –≤–µ—Ä–Ω–æ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

#### 4.1. –ü–æ–ª—É—á–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∞ –∏ –ø–æ–¥—Å—á—ë—Ç –æ—Å–Ω–æ–≤–Ω–æ–π –º–µ—Ç—Ä–∏–∫–∏

- –°—á–∏—Ç–∞–π—Ç–µ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `sessions_test` CSV-—Ñ–∞–π–ª —Å –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –æ —Å–µ—Å—Å–∏—è—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π `sessions_project_test.csv`.

- –í –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ `sessions_test` —Å–æ–∑–¥–∞–π—Ç–µ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü `good_session`. –í –Ω–µ–≥–æ –≤–æ–π–¥—ë—Ç –∑–Ω–∞—á–µ–Ω–∏–µ `1`, –µ—Å–ª–∏ –∑–∞ –æ–¥–Ω—É —Å–µ—Å—Å–∏—é –±—ã–ª–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–µ–Ω–æ 4 –∏ –±–æ–ª–µ–µ —Å—Ç—Ä–∞–Ω–∏—Ü, –∏ –∑–Ω–∞—á–µ–Ω–∏–µ `0`, –µ—Å–ª–∏ –ø—Ä–æ—Å–º–æ—Ç—Ä–µ–Ω–æ –º–µ–Ω—å—à–µ.

````python
sessions_test = pd.read_csv('https://code.s3.yandex.net/datasets/sessions_project_test.csv')
sessions_test['good_session'] = sessions_test['page_counter'].apply(lambda x: 1 if x >= 4 else 0)
sessions_test
````

#### 4.2. –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∞

–ü—Ä–µ–∂–¥–µ —á–µ–º –ø—Ä–∏—Å—Ç—É–ø–∞—Ç—å –∫ –∞–Ω–∞–ª–∏–∑—É –∫–ª—é—á–µ–≤—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤—ã—Ö –º–µ—Ç—Ä–∏–∫, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ —Ç–µ—Å—Ç –ø—Ä–æ–≤–µ–¥—ë–Ω –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∏ –≤—ã –±—É–¥–µ—Ç–µ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –¥–≤–µ —Å–æ–ø–æ—Å—Ç–∞–≤–∏–º—ã–µ –≥—Ä—É–ø–ø—ã.

- –†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–Ω—è –∏ –æ–±–µ–∏—Ö —Ç–µ—Å—Ç–æ–≤—ã—Ö –≥—Ä—É–ø–ø, –∏—Å–ø–æ–ª—å–∑—É—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É.

- –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –¥–Ω–µ–≤–Ω—ã—Ö —Å–µ—Å—Å–∏–π –≤ –¥–≤—É—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è –∏–ª–∏ —Ä–∞–∑–ª–∏—á–∏—è –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–º–∏. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª–∏—Ç —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫.

- –í –∫–∞—á–µ—Å—Ç–≤–µ –æ—Ç–≤–µ—Ç–∞ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ p-value –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–π—Ç–µ –µ–≥–æ.

<div style="background-color: #d4edda; color: #155724; padding: 15px; border-radius: 4px; border: 1px solid #c3e6cb;">

<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v1:</b>

–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã –≤–µ—Ä–Ω–æ. 

````python
#–†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–Ω—è –∏ –æ–±–µ–∏—Ö —Ç–µ—Å—Ç–æ–≤—ã—Ö –≥—Ä—É–ø–ø, –∏—Å–ø–æ–ª—å–∑—É—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É

session_group_a = sessions_test[sessions_test['test_group'] == 'A'].groupby('session_date').session_id.nunique().reset_index()
session_group_b = sessions_test[sessions_test['test_group'] == 'B'].groupby('session_date').session_id.nunique().reset_index()
````

````python
#–ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –¥–Ω–µ–≤–Ω—ã—Ö —Å–µ—Å—Å–∏–π –≤ –¥–≤—É—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è –∏–ª–∏ —Ä–∞–∑–ª–∏—á–∏—è 
#–Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã

plt.figure(figsize=(10, 8))

# –õ–∏–Ω–∏—è –¥–ª—è –≥—Ä—É–ø–ø—ã A
plt.plot(session_group_a['session_date'], 
         session_group_a['session_id'], 
         label='–ì—Ä—É–ø–ø–∞ A', 
         color='blue')

# –õ–∏–Ω–∏—è –¥–ª—è –≥—Ä—É–ø–ø—ã B
plt.plot(session_group_b['session_date'], 
         session_group_b['session_id'], 
         label='–ì—Ä—É–ø–ø–∞ B', 
         color='orange')  

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è
plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–µ—Å—Å–∏–π –ø–æ –¥–Ω—è–º –¥–ª—è –≥—Ä—É–ø–ø A –∏ B', fontsize=14)
plt.xlabel('–î–∞—Ç–∞', fontsize=12)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ—Å—Å–∏–π', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# –ü–æ–≤–æ—Ä–æ—Ç –¥–∞—Ç –Ω–∞ –æ—Å–∏ X –¥–ª—è –ª—É—á—à–µ–π —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏
plt.xticks(rotation=45)

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –ø–æ–¥–∂–∞—Ç–∏–µ layout
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫
plt.show()
````

````python
import scipy.stats as stats
````

````python
#–ò—Å–ø–æ–ª—å–∑—É–µ–º –¢-—Ç–µ—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª–∏—Ç —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫

session_group_ax = session_group_a['session_id']
session_group_bx = session_group_b['session_id']

t_stat, p_value = stats.ttest_ind(session_group_ax, session_group_bx, alternative='two-sided')
print(f'T-—Ç–µ—Å—Ç: t = {t_stat:.3f}, p-value = {p_value:.3f}')

alpha = 0.05

if p_value >= alpha:
    print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–∞—Ö–æ–¥–∏—Ç –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ! –í—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å—Ä–µ–¥–Ω–∏–µ –≤ –≥—Ä—É–ø–ø–∞—Ö A –∏ B —Ä–∞–≤–Ω—ã')
else:
    print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –ù–ï –Ω–∞—Ö–æ–¥–∏—Ç –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ! –í—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å—Ä–µ–¥–Ω–∏–µ –≤ –≥—Ä—É–ø–ø–∞—Ö A –∏ B –ù–ï —Ä–∞–≤–Ω—ã')
````

–û—Ç–ª–∏—á–Ω–∞—è —Ä–∞–±–æ—Ç–∞! –í–µ—Ä–Ω–æ –≤—ã–±—Ä–∞–Ω –∏ –ø—Ä–æ–≤–µ–¥–µ–Ω —Ç–µ—Å—Ç. –ü–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ–≥–ª–∞—Å–µ–Ω —Å –≤—ã–≤–æ–¥–∞–º–∏.

#### 4.3. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–ª–∏ —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π

–ö–æ–≥–¥–∞ –≤—ã —É–±–µ–¥–∏–ª–∏—Å—å, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ—Å—Å–∏–π –≤ –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö –Ω–µ —Ä–∞–∑–ª–∏—á–∞–ª–æ—Å—å, –º–æ–∂–Ω–æ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç—å –∫ –∞–Ω–∞–ª–∏–∑—É –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ ‚Äî –¥–æ–ª–∏ —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π.

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–æ–∑–¥–∞–Ω–Ω—ã–π –Ω–∞ –ø–µ—Ä–≤–æ–º —à–∞–≥–µ –∑–∞–¥–∞–Ω–∏—è —Å—Ç–æ–ª–±–µ—Ü `good_session` –∏ —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –¥–æ–ª—é —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è –≤—ã–±–æ—Ä–æ–∫ A –∏ B, –∞ —Ç–∞–∫–∂–µ —Ä–∞–∑–Ω–∏—Ü—É –≤ —ç—Ç–æ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ. –ü–æ–ª—É—á–µ–Ω–Ω—ã–π –≤—ã–≤–æ–¥ –æ—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω–µ.

````python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º –¥–æ–ª—é —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è –≤—ã–±–æ—Ä–æ–∫ A –∏ B, –∞ —Ç–∞–∫–∂–µ —Ä–∞–∑–Ω–∏—Ü—É –≤ —ç—Ç–æ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ

good_ration_a = round(100 * (sessions_test[(sessions_test.good_session == 1) & 
                            (sessions_test.test_group == 'A')].shape[0] / sessions_test[sessions_test.test_group == 'A'].shape[0]), 3)

good_ration_b = round(100 * (sessions_test[(sessions_test.good_session == 1) & 
                            (sessions_test.test_group == 'B')].shape[0] / sessions_test[sessions_test.test_group == 'B'].shape[0]), 3)

diff_b_a = round(good_ration_b - good_ration_a, 3)

print(f'–î–æ–ª—è —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è A = {good_ration_a}%')
print(f'–î–æ–ª—è —É—Å–ø–µ—à–Ω—ã—Ö —Å–µ—Å—Å–∏–π –¥–ª—è –í = {good_ration_b}%')
print(f'–†–∞–∑–Ω–∏—Ü–∞ B - A = {diff_b_a}%') 
````

#### 4.4. –ù–∞—Å–∫–æ–ª—å–∫–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏
