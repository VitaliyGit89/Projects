```python
#–£–±–µ—Ä–µ–º –≤—ã–±—Ä–æ—Å—ã –≤ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —ç—Ç–æ–≥–æ –≤—ã—á–∏—Å–ª–∏–º 1-–π –∏ 99-–π –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª–∏

q_min_msc = moscow['hours'].quantile(0.01)
q_max_msc = moscow['hours'].quantile(0.99)
msc_filtered = moscow[(moscow['hours'] >= q_min_msc) & (moscow['hours'] <= q_max_msc)]

q_min_spb = spb['hours'].quantile(0.01)
q_max_spb = spb['hours'].quantile(0.99)
spb_filtered = spb[(spb['hours'] >= q_min_spb) & (spb['hours'] <= q_max_spb)]

print(f'–†–∞–∑–º–µ—Ä –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –ú—Å–∫: {msc_filtered.shape[0]}')
print(f'–†–∞–∑–º–µ—Ä –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –°–ü–±: {spb_filtered.shape[0]}')
```

```python
#–ù–∞—Ö–æ–¥–∏–º –¥–∞—Ç—ã —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ (–º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è)
registrations = activity_df[activity_df['event_name'] == 'registration'].groupby('user_id')['event_dt'].min().reset_index().rename(columns={'event_dt': 'registration_dt'})

#–û–±—ä–µ–¥–∏–Ω–∏–º —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
df_with_reg = activity_df.merge(registrations, on='user_id', how='left')

#–î–æ–±–∞–≤–∏–º –∫–æ–ª–æ–Ω–∫—É —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏
df_with_reg['days_since_reg'] = round((df_with_reg['event_dt'] - df_with_reg['registration_dt']).dt.total_seconds() / 86400, 1)

#–û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ —Ç–µ —Å–æ–±—ã—Ç–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã –≤ —Ç–µ—á–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö —Å–µ–º–∏ –¥–Ω–µ–π —Å –º–æ–º–µ–Ω—Ç–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏
df_up_to_7 = df_with_reg[df_with_reg['days_since_reg'] <= 6]

df_up_to_7
```

- –±–∞–∑–æ–≤—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ ‚Äî 30%,

- –º–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞ ‚Äî 80%,

- –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞ ‚Äî 95%.

```python
# –ó–∞–¥–∞—ë–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
beta = 0.2  # –û—à–∏–±–∫–∞ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞, —á–∞—Å—Ç–æ 1 - –º–æ—â–Ω–æ—Å—Ç—å
power = 1 - beta  # –ú–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞
p = 0.3 # –ë–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –¥–æ–ª–∏
mde = 0.03 #* p  # –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –¥–µ—Ç–µ–∫—Ç–∏—Ä—É–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç
effect_size = proportion_effectsize(p, p + mde)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Å NormalIndPower
power_analysis = NormalIndPower()

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1 # –†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫
)

print(f"–ù–µ–æ–±—Ö–æ–¥–∏–º—ã–π —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã: {int(sample_size)}")
```
