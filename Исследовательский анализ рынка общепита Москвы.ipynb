### –ó–∞–¥–∞—á–∞ 7

–°–≥—Ä—É–ø–ø–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—è–º –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ –Ω–∞–π–¥–∏—Ç–µ —Ç–æ–ø-15 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —Å–µ—Ç–µ–π –≤ –ú–æ—Å–∫–≤–µ. –î–ª—è –Ω–∏—Ö –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞. –ü–æ–¥ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å—é –ø–æ–Ω–∏–º–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —ç—Ç–æ–π —Å–µ—Ç–∏ –≤ —Ä–µ–≥–∏–æ–Ω–µ. –ö –∫–∞–∫–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∑–∞–≤–µ–¥–µ–Ω–∏–π –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è? –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ–ø—Ä–æ–≤–æ–¥–∏—Ç–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–º–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è–º–∏.

```python
# –û—Ç–±–∏—Ä–∞–µ–º —Ç–æ–ª—å–∫–æ —Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è –∏ –≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é
df_chain = df_cleaned[df_cleaned['chain'] == 1]
df_chain_grouped = df_chain.groupby('name').size().reset_index(name='count')
```

```python
# –°–æ—Ä—Ç–∏—Ä—É–µ–º —Å–µ—Ç–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ –≤—ã–±–∏—Ä–∞–µ–º —Ç–æ–ø-15
top_15_chains = df_chain_grouped.sort_values(by='count', ascending=False).head(15)
top_15_chains
```

```python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é –∑–∞–≤–µ–¥–µ–Ω–∏—è –∏ —Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥
average_ratings = df_chain.groupby('name')['rating'].mean().reset_index(name='average_rating')
average_ratings
```

```python
# –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ —Å—Ä–µ–¥–Ω–µ–º —Ä–µ–π—Ç–∏–Ω–≥–µ
top_15_with_ratings = top_15_chains.merge(average_ratings, on='name')
top_15_with_ratings
```

```python
# –°–æ–∑–¥–∞–µ–º —Ç–∞–±–ª–∏—Ü—É —Ç–æ–ª—å–∫–æ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –∑–∞–≤–µ–¥–µ–Ω–∏–π –∏ –∏—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π –±–µ–∑ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
name_category_table = df[['name', 'category']].drop_duplicates()
```

```python
# –î–æ–±–∞–≤–ª—è–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é –∫ –æ—Å–Ω–æ–≤–Ω–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É 
top_15 = top_15_with_ratings.merge(name_category_table, on='name')
top_15
```

```python
# –ü–µ—Ä–≤—ã–π –≥—Ä–∞—Ñ–∏–∫: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π
plt.figure(figsize=(12, 6))
sns.barplot(data=top_15, x='name', y='count', palette='viridis')
plt.title('–¢–æ–ø-15 —Å–µ—Ç–µ–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–≤–µ–¥–µ–Ω–∏–π')
plt.xlabel('–ù–∞–∑–≤–∞–Ω–∏–µ —Å–µ—Ç–∏')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π')
plt.xticks(rotation=45)
plt.show()

# –í—Ç–æ—Ä–æ–π –≥—Ä–∞—Ñ–∏–∫: —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥
plt.figure(figsize=(12, 6))
sns.barplot(data=top_15, x='name', y='average_rating', palette='coolwarm')
plt.title('–°—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥ —Ç–æ–ø-15 —Å–µ—Ç–µ–π')
plt.xlabel('–ù–∞–∑–≤–∞–Ω–∏–µ —Å–µ—Ç–∏')
plt.ylabel('–°—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥')
plt.xticks(rotation=45)
plt.show()
```

```python
# –ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –±–æ–ª–µ–µ —Ä–µ–∞–ª–∏—Å—Ç–æ—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–ª–≤–µ–Ω–∏–µ —É–±–µ—Ä–µ–º –∞–Ω–æ–º–∞–ª—å–Ω–æ –≤—ã—Å–æ–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞
df_cleaned_avg_bill = df_cleaned.loc[df_cleaned['middle_avg_bill']<=2500]
```

```python
# –ò–∑—É—á–∏–º –≤–∞—Ä–∏–∞—Ü–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –∑–∞–≤–µ–¥–µ–Ω–∏—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–∞–π–æ–Ω–∞ –ú–æ—Å–∫–≤—ã
avg_bill = round(df_cleaned_avg_bill.groupby('district')['middle_avg_bill'].mean()).reset_index(name='avg_bill')
avg_bill_sorted = avg_bill.sort_values(by='avg_bill', ascending=False)
avg_bill_sorted
```
